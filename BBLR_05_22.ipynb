```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import Backend.NXCALS as nx

# Importing data
database = pd.read_parquet('Data/dataset_2022_05_22.parquet')
database.insert(0,'Time (s)',(database.index - database.index[0]).total_seconds())

MDWindow = [pd.Timestamp('2022-05-22 13:30'),pd.Timestamp('2022-05-22 16:45')]
MDTime = (MDWindow[1]-MDWindow[0]).total_seconds()

locations = ['L1B1','R1B2','L5B1','R5B2']
wires     = [nx.NXCALSWire(loc = loc) for loc in locations]
beams     = [nx.NXCALSBeam(name) for name in ['B1','B2']]
lineInfo  = [('b','-v'),('r','-v'),('b','-^'),('r','-^')]
```

### Tune VS Current

```python

%matplotlib inline
#%matplotlib widget

for beam in beams:
    
    plt.figure(figsize=(15, 8))
    plt.suptitle(beam.name)
    
    
    for wire,color in zip(wires,['C0','C0','C1','C1']):
        if wire.loc[-2:] != beam.name:
            continue
            
            
        toPlot = 'I'
        subset  = database.dropna(subset=[wire[toPlot]])
        plt.plot(subset.index,subset[wire[toPlot]],'-o',ms=2,label = wire.loc + '.' + wire.label[toPlot])
        
    plt.legend(loc='upper left')
    plt.xlabel('UTC Time')
    plt.ylabel(f"{wire.label[toPlot]} [{wire.units[toPlot]}]")
        
    ax2 = plt.gca().twinx()
    plt.sca(ax2)
    toPlot = 'Qx_HS'
    subset  = database.dropna(subset=[beam[toPlot]])
    plt.plot(subset.index,subset[beam[toPlot]],'.',color='C3',label = beam.label[toPlot])
    
    toPlot = 'Qy_HS'
    subset  = database.dropna(subset=[beam[toPlot]])
    plt.plot(subset.index,subset[beam[toPlot]],'.',color='C4',label = beam.label[toPlot])
    
    plt.legend(loc='upper right')
    plt.ylabel(r'$Q_{x,y}$')
    plt.xlim(MDWindow)
    
    
```
